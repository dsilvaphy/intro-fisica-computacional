Vamos escrever um jogo bem simples. Há um coelho na beira de um rio, fugindo de uma raposa. Para fugir, coelho vai saltar da margem até uma pedra a uma distância $d = 1{,}5\,\mathrm{m}$. A pedra tem uma largura $L = 30\,\mathrm{cm}$.

O jogador tem 5 vidas para tentar salvar o coelho. A cada tentativa, o jogo deve informar o número de vidas restante, e esperar o jogador digitar a velocidade em $\mathrm{cm\,s}^{-1}$ e o ângulo de lançamento em graus, usando a função `input()`.
